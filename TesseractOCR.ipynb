```python
# ==============================================
# 2. Importar librerías
# ==============================================
import cv2
import pytesseract
from PIL import Image
import numpy as np
from google.colab import files

# Configurar idioma (en Colab ya viene 'eng' instalado, se puede agregar 'spa')
lang = "eng"   # o "eng+spa" si instalás paquete español

# ==============================================
# 3. Subir una imagen desde tu PC
# ==============================================
uploaded = files.upload()

# Tomar el primer archivo subido
image_path = list(uploaded.keys())[0]
print("Imagen cargada:", image_path)

# ==============================================
# 4. Preprocesamiento básico
# ==============================================
img_bgr = cv2.imread(image_path)
img_gray = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2GRAY)
img_blur = cv2.medianBlur(img_gray, 3)
img_bin = cv2.adaptiveThreshold(img_blur, 255, cv2.ADAPTIVE_THRESH_MEAN_C,
                                cv2.THRESH_BINARY, 21, 10)

# ==============================================
# 5. OCR con Tesseract
# ==============================================
# psm 6 = bloque de texto uniforme
config = "--oem 3 --psm 6"

text = pytesseract.image_to_string(img_bin, lang=lang, config=config)
print("===== TEXTO DETECTADO =====")
print(text.strip())

# ==============================================
# 6. Extra: palabras con coordenadas y confianza
# ==============================================
data = pytesseract.image_to_data(img_bin, lang=lang, config=config, output_type=pytesseract.Output.DICT)

print("\n===== PALABRAS DETECTADAS =====")
n = len(data["text"])
for i in range(n):
    word = data["text"][i].strip()
    conf = int(data["conf"][i]) if data["conf"][i] != '-1' else -1
    if word != "" and conf >= 0:
        x, y, w, h = data["left"][i], data["top"][i], data["width"][i], data["height"][i]
        print(f"[{conf:3d}] '{word}' @ ({x},{y},{w},{h})")
```
