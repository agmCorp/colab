```python
# Paso 1: Instalar la librería necesaria
!pip install python-docx

import docx
import io
from google.colab import files

print("Por favor, sube tu archivo plantilla.docx:")

# Paso 2: Pedir al usuario que suba el archivo
uploaded = files.upload()

try:
    # Obtener el nombre y los bytes del archivo subido
    filename = next(iter(uploaded))
    file_bytes = uploaded[filename]
    print(f"\nArchivo '{filename}' cargado. Procesando...")

    # Abrir el documento desde los bytes en memoria
    doc = docx.Document(io.BytesIO(file_bytes))

    # Paso 3: Definir los reemplazos
    replacements = {
        "{{fecha}}": "2025/10/10",
        "{{hora}}": "13:00",
        "{{nombre_operario}}": "Álvaro Morales",

        # --- Descomenta y rellena los demás campos si es necesario ---
        # "{{linea}}": "Línea de Producción 5",
        # "{{nombre_producto}}": "Producto Ejemplo",
        # "{{ubicacion}}": "Almacén Central",
        # "{{comentarios_operario}}": "Sin observaciones.",
    }

    # Paso 4: Aplicar reemplazos en los párrafos principales
    for para in doc.paragraphs:
        for placeholder, value in replacements.items():
            if placeholder in para.text:
                para.text = para.text.replace(placeholder, value)

    # Paso 5: Aplicar reemplazos dentro de las tablas
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for para in cell.paragraphs:
                    for placeholder, value in replacements.items():
                        if placeholder in para.text:
                            para.text = para.text.replace(placeholder, value)

    print("Reemplazos aplicados exitosamente.")

    # --- INICIO DE LA CORRECCIÓN ---

    # Paso 6: Guardar el documento modificado directamente en el disco de Colab
    output_filename = 'resultado.docx'
    doc.save(output_filename)

    # --- FIN DE LA CORRECCIÓN ---

    # Paso 7: Descargar el archivo .docx que acabamos de guardar
    print(f"¡Proceso completado! Descargando '{output_filename}'...")
    files.download(output_filename) # Ahora 'resultado.docx' SÍ existe

except StopIteration:
    print("\nNo se subió ningún archivo.")
except Exception as e:
    print(f"\nOcurrió un error: {e}")
```

```python
# Paso 1: Instalar la librería para manipular DOCX
print("Instalando librerías...")
!pip install python-docx

import docx
import io
import os
from google.colab import files

print("Por favor, sube tu archivo plantilla.docx:")

# Paso 2: Pedir al usuario que suba el archivo
uploaded = files.upload()

try:
    # Obtener el nombre y los bytes del archivo subido
    filename = next(iter(uploaded))
    file_bytes = uploaded[filename]
    print(f"\nArchivo '{filename}' cargado. Procesando...")

    # Abrir el documento desde los bytes en memoria
    doc = docx.Document(io.BytesIO(file_bytes))

    # Paso 3: Definir los reemplazos
    replacements = {
        "{{fecha}}": "2025/10/10",
        "{{hora}}": "13:00",
        "{{nombre_operario}}": "Álvaro Morales",

        # OTRAS PLANTILLAS (Se han añadido valores por defecto para evitar errores)
        # Puedes cambiar estos si lo necesitas:
        "{{linea}}": "Línea 1",
        "{{nombre_producto}}": "Lote NNN",
        "{{ubicacion}}": "Almacén Oeste",
        "{{comentarios_operario}}": "Sin observaciones.",
    }

    # Paso 4: Aplicar reemplazos en el contenido (párrafos y tablas)
    # Reemplazos en párrafos principales
    for para in doc.paragraphs:
        for placeholder, value in replacements.items():
            if placeholder in para.text:
                para.text = para.text.replace(placeholder, value)

    # Reemplazos dentro de tablas
    for table in doc.tables:
        for row in table.rows:
            for cell in row.cells:
                for para in cell.paragraphs:
                    for placeholder, value in replacements.items():
                        if placeholder in para.text:
                            para.text = para.text.replace(placeholder, value)

    print("Reemplazos aplicados exitosamente.")

    # Nombres de archivo
    input_filename = 'resultado.docx'
    output_filename = 'resultado.pdf'

    # Paso 5: Guardar el documento modificado en DOCX
    doc.save(input_filename)

    # -----------------------------------------------------------------
    # Paso 6: INSTALACIÓN Y CONVERSIÓN A PDF
    # -----------------------------------------------------------------
    print("\nInstalando LibreOffice para la conversión a PDF...")
    # Instalar LibreOffice (necesario en Linux/Colab para esta conversión)
    !apt-get update > /dev/null
    !apt-get install -y libreoffice > /dev/null

    print("Convirtiendo DOCX a PDF...")
    # Comando de shell para convertir DOCX a PDF usando LibreOffice en modo "headless"
    # --outdir /content/ asegura que el PDF se cree en el directorio actual
    !libreoffice --headless --convert-to pdf '{input_filename}' --outdir /content/

    # Comprobar si el archivo PDF fue creado
    if not os.path.exists(output_filename):
        raise FileNotFoundError(f"El archivo {output_filename} no se pudo generar. Verifique la plantilla.")

    # Paso 7: Descargar el nuevo archivo PDF
    print(f"¡Proceso completado! Descargando '{output_filename}'...")
    files.download(output_filename)

except StopIteration:
    print("\nNo se subió ningún archivo.")
except FileNotFoundError as e:
    print(f"\nOcurrió un error al guardar o convertir el archivo: {e}")
except Exception as e:
    print(f"\nOcurrió un error inesperado: {e}")
```
