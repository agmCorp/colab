```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Cargar la imagen original en color (BGR)
original_image = cv2.imread('/content/1316.png')
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convertir de BGR a RGB para matplotlib

# Aplicar corrección gamma
gamma = 0.3
gamma_corrected = np.array(255 * (original_image / 255) ** gamma, dtype='uint8')

# Aplicar filtrado mediano
processed_image = cv2.medianBlur(gamma_corrected, ksize=3)  # ksize puede ser 3 o 5

# Guardar la imagen procesada en tamaño original
cv2.imwrite('/content/1316_modificada.png', cv2.cvtColor(processed_image, cv2.COLOR_RGB2BGR))
print("Imagen guardada como /content/1316_modificada.png")

# Mostrar la imagen original y la mejorada
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(original_image)
axs[0].axis('off')

axs[1].imshow(processed_image)
axs[1].axis('off')

plt.tight_layout()
plt.show()

# -----------------------------
# AHORA MOSTRAR LOS HISTOGRAMAS
# -----------------------------

# Convertir las imágenes a escala de grises para calcular los histogramas de luminancia
original_gray = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)
processed_gray = cv2.cvtColor(processed_image, cv2.COLOR_RGB2GRAY)

# Crear un gráfico para los histogramas
fig2, axs2 = plt.subplots(1, 2, figsize=(10, 5))

# Histograma de la imagen original
axs2[0].hist(original_gray.ravel(), bins=256, color='gray', alpha=0.9)
axs2[0].set_xlabel("Intensidad")
axs2[0].set_ylabel("Número de Píxeles")
axs2[0].set_xlim(0, 255)

# Histograma de la imagen mejorada
axs2[1].hist(processed_gray.ravel(), bins=256, color='gray', alpha=0.9)
axs2[1].set_xlabel("Intensidad")
axs2[1].set_ylabel("Número de Píxeles")
axs2[1].set_xlim(0, 255)

plt.tight_layout()
plt.show()

# -----------------------------
# PSNR y SSIM
# -----------------------------

# Calcular PSNR
valor_psnr = psnr(original_gray, processed_gray)

# Calcular SSIM
valor_ssim, _ = ssim(original_gray, processed_gray, full=True)

# Mostrar resultados
print("PSNR:", valor_psnr)
print("SSIM:", valor_ssim)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.metrics import structural_similarity as ssim

# Cargar la imagen original en color (BGR)
original_image = cv2.imread('/content/4285.png')
original_image = cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB)  # Convertir a RGB para matplotlib

# Convertir a YCrCb para trabajar sobre luminancia
ycrcb_image = cv2.cvtColor(original_image, cv2.COLOR_RGB2YCrCb)

# Separar canales
y_channel, cr_channel, cb_channel = cv2.split(ycrcb_image)

# Aplicar CLAHE sobre el canal Y
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))  # Parámetros estándar de CLAHE
y_clahe = clahe.apply(y_channel)

# Recombinar canales
ycrcb_clahe = cv2.merge((y_clahe, cr_channel, cb_channel))
clahe_image = cv2.cvtColor(ycrcb_clahe, cv2.COLOR_YCrCb2RGB)

# Aplicar filtro de media (promedio) sobre la imagen mejorada
processed_image = cv2.blur(clahe_image, (3, 3))  # Filtro de media 3x3

# Guardar la imagen procesada en tamaño original
cv2.imwrite('/content/4285_modificada.png', cv2.cvtColor(processed_image, cv2.COLOR_RGB2BGR))
print("Imagen guardada como /content/4285_modificada.png")

# Mostrar la imagen original y la mejorada
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

axs[0].imshow(original_image)
axs[0].axis('off')

axs[1].imshow(processed_image)
axs[1].axis('off')

plt.tight_layout()
plt.show()

# -----------------------------
# AHORA MOSTRAR LOS HISTOGRAMAS
# -----------------------------

# Convertir las imágenes a escala de grises para calcular los histogramas de luminancia
original_gray = cv2.cvtColor(original_image, cv2.COLOR_RGB2GRAY)
processed_gray = cv2.cvtColor(processed_image, cv2.COLOR_RGB2GRAY)

# Crear un gráfico para los histogramas
fig2, axs2 = plt.subplots(1, 2, figsize=(10, 5))

# Histograma de la imagen original
axs2[0].hist(original_gray.ravel(), bins=256, color='gray', alpha=0.9)
axs2[0].set_xlabel("Intensidad")
axs2[0].set_ylabel("Número de Píxeles")
axs2[0].set_xlim(0, 255)

# Histograma de la imagen mejorada
axs2[1].hist(processed_gray.ravel(), bins=256, color='gray', alpha=0.9)
axs2[1].set_xlabel("Intensidad")
axs2[1].set_ylabel("Número de Píxeles")
axs2[1].set_xlim(0, 255)

plt.tight_layout()
plt.show()

# -----------------------------
# PSNR y SSIM
# -----------------------------

# Calcular PSNR
valor_psnr = psnr(original_gray, processed_gray)

# Calcular SSIM
valor_ssim, _ = ssim(original_gray, processed_gray, full=True)

# Mostrar resultados
print("PSNR:", valor_psnr)
print("SSIM:", valor_ssim)
```
